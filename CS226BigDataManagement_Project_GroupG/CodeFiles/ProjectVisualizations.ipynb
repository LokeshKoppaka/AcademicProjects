```python
# Vegas Heat Map!
import ast
import folium
import folium.plugins as plugins
#logic to read file and retrieve latitude and longitude 
# Note - Please ensure to change the file path!
with open('d:/UCR/UCR_FirstQuater/Big Data/BigData_Project/projectResults/vegas_heat_map.txt','r') as myfile:
    data = myfile.read().replace('\n','')
# Logic to convert string of list to list    
heatMapVal = ast.literal_eval(data)
lat = 36.127430
lon = -115.138460
# Setting plotting attributes - Folium
zoom_start=11
m = folium.Map(location=[lat, lon], tiles="OpenStreetMap", zoom_start=zoom_start)
hm = plugins.HeatMapWithTime(heatMapVal,max_opacity=0.3,auto_play=True,display_index=True,radius=7)
hm.add_to(m)
m
```

```python
# Scatter Plots for Las Vegas & Phoenix
import numpy as np
import matplotlib.pyplot as plt

#logic to read file and retrieve latitude and longitude 
lat_vegas = []
lon_vegas = []
lat_pheonix = []
lon_pheonix = []
# Logic for Las Vegas
# Note - Please ensure to change the file path!
with open("d:/UCR/UCR_FirstQuater/Big Data/BigData_Project/projectResults/vegas_geo_tag.csv", "r") as file:
    lis = [line.replace('\n', '').split(',') for line in file]
    for i,x in enumerate(lis):        
        if x[4] == "true":
            lat_vegas.append(float(x[1]))
            lon_vegas.append(float(x[2]))
# Logic for Phoenix
# Note - Please ensure to change the file path!
with open("d:/UCR/UCR_FirstQuater/Big Data/BigData_Project/projectResults/phoenix_geo_tag.csv", "r") as file:
    lis = [line.replace('\n', '').split(',') for line in file]
    for i,x in enumerate(lis):        
        if x[4] == "true":
            lat_pheonix.append(float(x[1]))
            lon_pheonix.append(float(x[2]))

            
#Logic to map them to scatter plot - Vegas
area = np.pi*0.05
plt.scatter(lon_vegas,lat_vegas, s=area, alpha=0.5)
plt.title("Vegas!!!",fontsize=16);
plt.xlabel('Longitude', fontsize=12)
plt.ylabel('Latitude', fontsize=12)
plt.show()
#Logic to map them to scatter plot - Phoenix
plt.scatter(lon_pheonix,lat_pheonix, s=area , alpha=0.5)
plt.title("Phoenix!!!", fontsize=16);
plt.xlabel('Longitude', fontsize=12)
plt.ylabel('Latitude', fontsize=12)
plt.show()
```

```python
#Rating distribution
import numpy as np
import matplotlib.pyplot as plt

#logic to read file and retrieve Stars and respective Count 
stars = []
counts = []
# Note - Please ensure to change the file path!
with open("d:/UCR/UCR_FirstQuater/Big Data/BigData_Project/projectResults/rating_distribution.csv", "r") as file:
    lis = [line.replace('\n', '').split(',') for line in file]
    for i,x in enumerate(lis):        
        stars.append(float(x[0]))
        counts.append(float(x[1]))
# Setting plotting attributes
xAxis = np.arange(len(stars))
plt.title("Star Ratings distribution", fontsize = 16)
plt.bar(xAxis, height = counts)
plt.xticks(xAxis, stars)
plt.xlabel('Star Ratings', fontsize=12)
plt.ylabel('Count', fontsize=12)
```

```python
#Popular Business Categories
import numpy as np
import matplotlib.pyplot as plt

#logic to read file and retrieve categories and respective counts 
categories = []
counts = []
# Note - Please ensure to change the file path!
with open("d:/UCR/UCR_FirstQuater/Big Data/BigData_Project/projectResults/top_categories.csv", "r") as file:
    lis = [line.replace('\n', '').split(',') for line in file]
    for i,x in enumerate(lis):        
        categories.append(x[0])
        counts.append(float(x[1]))
# Setting plotting attributes
xAxis = np.arange(len(categories[:20]))
plt.bar(xAxis, height = counts[:20])
plt.xticks(xAxis, categories[:20], rotation="vertical")
plt.title("Top Businesses in United States!", fontsize = 16)
plt.ylabel('Count', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
#Top cities with business review in USA - which city has the most review!
import numpy as np
import matplotlib.pyplot as plt

#logic to read file and retrieve cities and respective counts 
cities = []
counts = []
# Note - Please ensure to change the file path!
with open("d:/UCR/UCR_FirstQuater/Big Data/BigData_Project/projectResults/cities_with_most_reviews.csv", "r") as file:
    lis = [line.replace('\n', '').split(',') for line in file]
    for i,x in enumerate(lis):        
        cities.append(x[0])
        counts.append(float(x[1]))
# Setting plotting attributes
xAxis = np.arange(len(cities[:20]))
plt.bar(xAxis, height = counts[:20])
plt.xticks(xAxis, cities[:20], rotation="vertical")
plt.title("Top Cities", fontsize = 16)
plt.ylabel('Business Count', fontsize=12)
plt.tight_layout()
plt.show()
```
